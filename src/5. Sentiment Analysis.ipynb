```python
df_test = pd.read_csv("/content/drive/Shareddrives/TextAnalytics_project/dataset_ironita_2018/Data/processed_text_test.csv")
df_train = pd.read_csv("/content/drive/Shareddrives/TextAnalytics_project/dataset_ironita_2018/Data/processed_text_train.csv")
```

```python
df = pd.concat([df_train, df_test])
```

# FEEL-IT classifier

The feel-it-italian-emotion model performs emotion classification (joy, fear, anger, sadness) on Italian. It has fine-tuned the UmBERTo model on our new dataset (i.e., FEEL-IT) obtaining state-of-the-art performances on different benchmark corpora.

From https://huggingface.co/MilaNLProc/feel-it-italian-emotion

```python
emotion_classifier = EmotionClassifier()
sentiment_classifier = SentimentClassifier()
```

```python
tweets_list = []
for index, row in df.iterrows():
  tweets_list.append(row[-1])
```

```python
df["sentiment"] = sentiment_classifier.predict(tweets_list)
```

```python
df["emotion"] = emotion_classifier.predict(tweets_list)
```
